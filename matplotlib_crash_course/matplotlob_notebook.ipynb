# Basic Graph
## Basic Documentation @ https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html

```python
from matplotlib import pyplot as plt

x = [1, 2, 3, 4]
y = [2, 5, 4, 1]
plt.plot(x, y)
plt.title("Graph v002", fontdict={"fontname": "Helvetica", "fontsize": 20})
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Basic Graph with Custom Ticks and Legend

```python
x = [1, 2, 3, 4]
y = [2, 5, 4, 1]
plt.plot(x, y, label="Random")
plt.title("Graph v002", fontdict={"fontname": "Helvetica", "fontsize": 20})
plt.xlabel("X")
plt.ylabel("Y")
plt.xticks([1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.legend()
plt.show()
```

# Basic Graph with Custom Ticks, Legend, Color, Style and Two Lines

```python
x = [1, 2, 3, 4]
y = [2, 5, 4, 1]
plt.plot(x, y, label="Random", color="green", marker=".", markersize=12, markeredgecolor="orange", linestyle="--")
plt.plot([1,2,3,4], [1,2,3,4])
plt.title("Graph v003", fontdict={"fontname": "Helvetica", "fontsize": 20})
plt.xlabel("X")
plt.ylabel("Y")
plt.xticks([1, 2, 3, 4])
plt.yticks([2, 4, 6, 8])
plt.legend()
plt.show()
```

# Bar Charts

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]
bars = plt.bar(labels, values)
bars[0].set_hatch("/")
bars[1].set_hatch("*")
plt.show()
```

# Example of Graph With Real Data

```python
import pandas as pd
import numpy as np

gas_data = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8, 5))

plt.title("Price of Gas Over Years")
plt.xlabel("Years")
plt.ylabel("Price of Gas")

canada_plot = plt.plot(gas_data.Year, gas_data.Canada, label="Canada", color="red")
australia_plot = plt.plot(gas_data.Year, gas_data.Australia, label="Australia", color="Blue")

# years = [year for year in range(1990, 2008+1, 2)]
plt.xticks(gas_data.Year[::2])

plt.legend()

plt.show()
```

# Load FIFA Data -> Histogram

```python
fifa_data = pd.read_csv("fifa_data.csv")

bins = [num for num in range(40, 101, 10)]

plt.hist(fifa_data.Overall, bins=bins)

plt.xlabel("Skill of Players")
plt.ylabel("Number of Players")

plt.yticks()

plt.show()
```
